```python
from pytrends.request import TrendReq

pytrends = TrendReq(hl='zh-TW', tz=-480)
product_searches = dict()

for names in tqdm(products):
    kw_list = [names]
    pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='TW', gprop='')
    search_data = pytrends.interest_over_time()
    if len(search_data) > 0:
        product_searches.update(search_data.to_dict())
    else:
        print(names)
    
    sleep_sec = np.random.randint(3, 5)
    time.sleep(sleep_sec)

```
